# 1. 蒙特卡洛模拟

蒙特卡洛方法也称为计算机随机模拟方法，是基于对大量事件的统计结果来实现--些确定性问题的计算。

尽管整数规划由于限制变量为整数而增加了难度;然而又由于整数解是有限个,于是为枚举法提供了方便。
当然,在自变量维数很大和取值范围很宽的情况下,企图用显枚举法(即穷举法)计算出最优值是不现实的,但是应用概率理论可以证明,在一定
计算量的情况下,用蒙特卡洛法完全可以得出一个满意解。

## 例1



在python代码中，通过对模拟总数分别为100，1000，10000，100000进行100次模拟实验，并画出频次图。

```python
count = 100
times = [100, 1000, 10000, 100000]
for ele in times:
    for per in range(count):
        frequence = 0
        for time in range(1, ele):
            x = random.uniform(0, 12)
            y = random.uniform(0, 9)
            if (y < x**2 and x <= 3) or (y < 12 - x and x >= 3):
                frequence += 1
        area_appr = 12 * 9 * frequence / ele
        if ele == times[0]:
            plt.plot(times[0], area_appr, 'b*')
        if ele == times[1]:
            plt.plot(times[1], area_appr, 'g1')
        if ele == times[2]:
            plt.plot(times[2], area_appr, 'ro')
        if ele == times[3]:
            plt.plot(times[3], area_appr, 'dk')
plt.show()
```

可以看出模拟总数越大，频次越聚集——大概在49.5左右
